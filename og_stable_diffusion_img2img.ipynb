```python
import os
from diffusers import StableDiffusionImg2ImgPipeline

model_id = "runwayml/stable-diffusion-v1-5"
MODEL_CACHE = "diffusers-cache"

os.makedirs(MODEL_CACHE, exist_ok=True)

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    model_id,
    cache_dir=MODEL_CACHE,
).to("cuda")
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
from PIL import Image
import torch

# Define the function to run the model
def run_prediction(input_image_path, prompt="Van Gogh style"):
    init_img = Image.open(input_image_path)
    init_img = init_img.resize((800, 600))

    output = pipe(prompt=prompt, image=init_img, strength= 0.6, guidance_scale=10.0)

    output_image = output.images[0]
    output_image.save("/content/drive/MyDrive/CycleGan/vpkhoa_out.jpg")
    return output_image

# Example usage
output_image = run_prediction("/content/drive/MyDrive/CycleGan/vpkhoa.jpg")
output_image.show()
```
